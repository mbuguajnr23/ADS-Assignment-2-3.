```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
grouped = df.groupby(["JobRole","Attrition"])
```

```python
mean_distance = grouped["DistanceFromHome"].mean()
```

```python
mean_distance = mean_distance.reset_index()
```

```python
plt.bar(mean_distance["JobRole"],
mean_distance["DistanceFromHome"])
plt.xlabel("Job Role")
plt.xticks(rotation = 90)
plt.ylabel("Mean Distance From Home")
plt.title("Distance From Home by Job Role and Attrition")
plt.show()
```

```python
# Group the data by education and attrition
grouped = df.groupby(["Education", "Attrition"])
```

```python
# Calculate the average monthly income for each group
avg_income = grouped["MonthlyIncome"].mean()
```

```python
# Pivot the data so that the education levels are the columns and the attrition status is the index
pivoted = avg_income.unstack()
```

```python
# Plot the data as a bar chart
pivoted.plot(kind="bar")
```
